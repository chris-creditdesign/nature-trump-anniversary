## Remove the negative, zero or missing numbers

```python
len(df_nih)
```

```python
# From df_nih drop any row were 'estimated_remaning' is equal or less than zero or NaN
df_nih_filtered = df_nih[df_nih['estimated_remaining'] > 0]
```

```python
len(df_nih_filtered)
```

```python
if len(df_nih_filtered) == len(df_nih):
    print("No rows were dropped.")
else:
    print(f"{len(df_nih) - len(df_nih_filtered)} rows were dropped.")
```

```python
len(df_nsf)
```

```python
df_nsf_filtered = df_nsf[df_nsf['estimated_remaining'] > 0]
```

```python
len(df_nsf_filtered)
```

```python
if len(df_nsf_filtered) == len(df_nsf):
    print("No rows were dropped.")
else:
    print(f"{len(df_nsf) - len(df_nsf_filtered)} rows were dropped.")
```

## Join the two dataframes together

```python
df_combined = pd.concat([df_nih_filtered, df_nsf_filtered], ignore_index=True)
```

### Length of Combined DataFrame
