# NIH Grant Data Quality Analysis

This notebook analyzes NIH grant CSV files to identify and export grants with data quality issues:
1. Grants with zero or null 'Total Cost' amounts
2. Grants with empty or missing 'Project Terms'

The analysis processes CSV files from 2015-2025 and creates CSV reports for data quality assessment.

```python
import pandas as pd
from tqdm import tqdm
import os
from pathlib import Path


# Set display options for better readability
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 100)
```

```python
file_paths = [
    'SearchResult_Export_19Dec2025_065513.csv',
    'SearchResult_Export_19Dec2025_070252.csv',
    'SearchResult_Export_19Dec2025_070216.csv',
    'SearchResult_Export_19Dec2025_070140.csv',
    'SearchResult_Export_19Dec2025_070059.csv',
    'SearchResult_Export_19Dec2025_070020.csv',
    'SearchResult_Export_19Dec2025_065939.csv',
    'SearchResult_Export_19Dec2025_065857.csv',
    'SearchResult_Export_19Dec2025_065814.csv',
    'SearchResult_Export_19Dec2025_065732.csv',
    'SearchResult_Export_19Dec2025_065651.csv'
]
```

```python
# Load and combine all CSV files into one DataFrame
data_frames = []

# Process each CSV file
for file_path in tqdm(file_paths, desc="Loading CSV files"):
    try:
        # Read CSV file, skipping first 6 lines of metadata
        full_path = f'data/nih/{file_path}'
        df = pd.read_csv(full_path, skiprows=6)
        
        # Keep only the required columns
        required_columns = ["Project Terms", "Project Title", "Project Number", "Fiscal Year", "Total Cost"]
        df_filtered = df[required_columns].copy()
        
        # Add to list
        data_frames.append(df_filtered)
        print(f"Loaded {file_path}: {len(df_filtered)} rows")
        
    except FileNotFoundError:
        print(f"Warning: File {file_path} not found")
    except KeyError as e:
        print(f"Warning: Column not found in {file_path}: {e}")
    except Exception as e:
        print(f"Error loading {file_path}: {e}")

# Combine all DataFrames
if data_frames:
    combined_df = pd.concat(data_frames, ignore_index=True)
    
    # Sort by Fiscal Year
    combined_df = combined_df.sort_values('Fiscal Year')
    
    print(f"\nCombined DataFrame:")
    print(f"Total rows: {len(combined_df)}")
    print(f"Columns: {list(combined_df.columns)}")
    print(f"Fiscal Year range: {combined_df['Fiscal Year'].min()} - {combined_df['Fiscal Year'].max()}")
    
    # Display first few rows
    print(f"\nFirst 5 rows:")
    print(combined_df.head())
    
    # Display data types and basic info
    print(f"\nDataFrame info:")
    combined_df.info()
    
else:
    print("No data was loaded successfully.")
```

## Filter Grants with Zero, Null, or Missing Total Cost

```python
# Filter grants with zero, null, or missing Total Cost
print("Filtering grants with zero, null, or missing Total Cost...")

# Convert Total Cost to numeric, handling any string values
combined_df['Total Cost'] = pd.to_numeric(combined_df['Total Cost'], errors='coerce')

# Create filter for zero, null, or missing Total Cost
zero_null_cost_filter = (
    (combined_df['Total Cost'].isna()) |
    (combined_df['Total Cost'] == 0)
)

df_zero_null_cost = combined_df[zero_null_cost_filter].copy()

print(f"Found {len(df_zero_null_cost)} grants with zero, null, or missing Total Cost")
print(f"Percentage of total grants: {len(df_zero_null_cost) / len(combined_df) * 100:.3f}%")

if len(df_zero_null_cost) > 0:
    print("\nBreakdown by Fiscal Year:")
    year_breakdown = df_zero_null_cost['Fiscal Year'].value_counts().sort_index()
    for year, count in year_breakdown.items():
        print(f"  {year}: {count} grants")
    
    print("\nSample of grants with zero/null/missing Total Cost:")
    print(df_zero_null_cost.head())
else:
    print("No grants found with zero, null, or missing Total Cost.")
```

## Filter Grants with Empty or Missing Project Terms

```python
# Filter grants with empty or missing Project Terms
print("Filtering grants with empty or missing Project Terms...")

# Create filter for empty or missing Project Terms
empty_terms_filter = (
    (combined_df['Project Terms'].isna()) |
    (combined_df['Project Terms'] == '') |
    (combined_df['Project Terms'].astype(str).str.strip() == '') |
    (combined_df['Project Terms'].astype(str).str.strip() == 'nan')
)

df_empty_terms = combined_df[empty_terms_filter].copy()

print(f"Found {len(df_empty_terms)} grants with empty or missing Project Terms")
print(f"Percentage of total grants: {len(df_empty_terms) / len(combined_df) * 100:.3f}%")

if len(df_empty_terms) > 0:
    print("\nBreakdown by Fiscal Year:")
    year_breakdown = df_empty_terms['Fiscal Year'].value_counts().sort_index()
    for year, count in year_breakdown.items():
        print(f"  {year}: {count} grants")
    
    print("\nSample of grants with empty Project Terms:")
    print(df_empty_terms.head())
else:
    print("No grants found with empty or missing Project Terms.")
```

## Create Export Directory Structure

```python
# Create export directory structure
print("Setting up export directory structure...")

# Create export/nih directory if it doesn't exist
export_dir = Path('./export/nih')
export_dir.mkdir(parents=True, exist_ok=True)

print(f"‚úì Export directory created/verified: {export_dir}")
print(f"  Directory exists: {export_dir.exists()}")
print(f"  Directory is writable: {os.access(export_dir, os.W_OK)}")

# Define output file paths
zero_cost_file = export_dir / 'nih_grants_zero_null_cost.csv'
empty_terms_file = export_dir / 'nih_grants_empty_terms.csv'

print("\nPlanned output files:")
print(f"  1. Zero/null/missing Total Cost: {zero_cost_file}")
print(f"  2. Empty Project Terms: {empty_terms_file}")
```

## Export Filtered Data to CSV Files

```python
# Export filtered data to CSV files
print("Exporting filtered data to CSV files...")

# Define columns to export (all available columns)
export_columns = ["Project Terms", "Project Title", "Project Number", "Fiscal Year", "Total Cost"]

# Export grants with zero, null, or missing Total Cost
if len(df_zero_null_cost) > 0:
    df_zero_null_cost[export_columns].to_csv(zero_cost_file, index=False)
    print(f"‚úì Exported {len(df_zero_null_cost)} grants with zero/null/missing Total Cost to:")
    print(f"  {zero_cost_file}")
else:
    print("‚úì No grants with zero/null/missing Total Cost to export")

# Export grants with empty Project Terms
if len(df_empty_terms) > 0:
    df_empty_terms[export_columns].to_csv(empty_terms_file, index=False)
    print(f"‚úì Exported {len(df_empty_terms)} grants with empty Project Terms to:")
    print(f"  {empty_terms_file}")
else:
    print("‚úì No grants with empty Project Terms to export")

print("\nExport completed successfully!")
```

## Display Summary Statistics

```python
# Display comprehensive summary statistics
print("=" * 60)
print("NIH GRANT DATA QUALITY ANALYSIS SUMMARY")
print("=" * 60)

print("\nüìä OVERALL STATISTICS:")
print(f"  ‚Ä¢ Total grants processed: {len(combined_df):,}")
print(f"  ‚Ä¢ Fiscal Years covered: {combined_df['Fiscal Year'].min()}-{combined_df['Fiscal Year'].max()}")
# Calculate total funding, excluding null/zero values
valid_costs = combined_df['Total Cost'].dropna()
valid_costs = valid_costs[valid_costs > 0]
print(f"  ‚Ä¢ Total valid funding amount: ${valid_costs.sum():,.2f}")
print(f"  ‚Ä¢ Average grant amount: ${valid_costs.mean():,.2f}")

print("\nüîç DATA QUALITY ISSUES:")
print(f"  ‚Ä¢ Grants with zero/null/missing Total Cost: {len(df_zero_null_cost):,} ({len(df_zero_null_cost)/len(combined_df)*100:.3f}%)")
print(f"  ‚Ä¢ Grants with empty/missing Project Terms: {len(df_empty_terms):,} ({len(df_empty_terms)/len(combined_df)*100:.3f}%)")

# Show breakdown by Fiscal Year
print("\nüìÖ GRANTS BY FISCAL YEAR:")
year_stats = combined_df.groupby('Fiscal Year').agg({
    'Project Number': 'count',
    'Total Cost': ['sum', 'mean']
}).round(2)
year_stats.columns = ['Count', 'Total_Funding', 'Mean_Amount']
print(year_stats)

# Show data quality issues by year
print("\nüìâ DATA QUALITY ISSUES BY FISCAL YEAR:")
quality_issues = pd.DataFrame({
    'Zero_Null_Cost': df_zero_null_cost['Fiscal Year'].value_counts().sort_index(),
    'Empty_Terms': df_empty_terms['Fiscal Year'].value_counts().sort_index()
}).fillna(0).astype(int)
print(quality_issues)

print("\nüìÅ EXPORTED FILES:")
if len(df_zero_null_cost) > 0:
    print(f"  ‚úì {zero_cost_file.name}")
if len(df_empty_terms) > 0:
    print(f"  ‚úì {empty_terms_file.name}")

print("\nüéâ Analysis completed successfully!")
print("=" * 60)
```
